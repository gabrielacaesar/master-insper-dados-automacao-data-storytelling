```python
import pandas as pd
import altair as alt
```

```python
marathon = pd.read_csv("nym2002.csv")
```

```python
marathon.groupby('gender')['time'].describe()
```

```python
marathon.corr() # para ver a correcao
```

```python
# exercicio 4
```

```python
bins = [0, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80, 100]
```

```python
marathon['age_group'] = pd.cut(marathon['age'], bins = bins) # como se fosse um case_when()
```

```python
marathon.head()
```

```python
marathon.info() # observe que age_group ficou como categoria
```

```python
marathon['age_group'] = marathon['age_group'].astype('str') # de categoria para string
```

```python
marathon.info()
```

```python
alt.Chart(marathon).mark_boxplot().encode(
    x = 'gender',
    y = 'time',
    column = 'age_group'
)
```

```python
alt.Chart(marathon).mark_point().encode(
    x = 'age',
    y = 'time',
    color  = 'age_group',
    column = 'gender'
)
```
