```python
# Primeiro, listamos os arquivos que queremos ler
# Antes de escrever o código, à esquerda, fazemos o upload dos arquivos txt
# Depois, rodamos o código abaixo
list_files = ("GMT20210726-212633_Recording.txt", "GMT20210728-223542_Recording.txt", 
              "GMT20210802-212614_Recording.txt", "GMT20210804-222232_Recording.txt", 
              "GMT20210806-222707_Recording.txt", "GMT20210809-212840_Recording.txt", 
              "GMT20210809-213308_Recording.txt", "GMT20210809-234715_Recording.txt", 
              "GMT20210811-223050_Recording.txt", "GMT20210813-222825_Recording.txt", 
              "GMT20210816-212317_Recording.txt", "GMT20210816-234945_Recording.txt", 
              "GMT20210818-222332_Recording.txt", "GMT20210820-222253_Recording.txt")
```

```python
# Agora, vamos criar uma iteração para ler os arquivos
# Depois, vamos ler as linhas e, em seguida, as partes
# Por fim, vamos filtrar para manter apenas mensagens com 'http'

for arquivo_txt in list_files: # para cada arquivo txt listado vamos fazer o processo abaixo dentro do loop
  #print(arquivo_txt) 
  arquivo = open(arquivo_txt) # cada arquivo será aberto
  for linha in arquivo: # para cada linha do arquivo fazemos o processo abaixo dentro de mais um loop
      partes = linha.strip().split("\t") # em cada arquivo lido precisamos organizar as linhas
      #print(partes)
      if len(partes) == 3: # quando forem 3 itens, vamos separar para manter cada texto em um campo
        mensagem = partes[2].strip() # para selecionar apenas as mensagens, e não nome e hora
        #print(mensagem)
      if 'http' in mensagem: # condição para manter apenas quando tiver 'http'
        print(mensagem)
```
