```python
  import requests

  def verbete_resumido(titulo):
    url = 'https://pt.wikipedia.org/api/rest_v1/page/summary/' + titulo
    resposta = requests.get(url)
    dados = resposta.json()
    return dados['extract']
```

```python
print(verbete_resumido('Python'))
```

```python
print(verbete_resumido('Google'))
```

```python
print(verbete_resumido('Terminal'))
```
